# Task-01:Import Data Set

```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, accuracy_score
```

```python
df = pd.read_csv('weight-height.csv')
df.head()
```

# Task-02:Separate X (Gender, Height) and Y (y=Weight).

```python
# Step 2: Separate X (features) and Y (target)
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})  # Encoding Gender
X = df[['Gender', 'Height']]
Y = df['Weight']
```

# Task-03:Train = 70%, Test = 30%

```python
# Step 3: Train-Test Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state=42)
```

# Task-04: Apply Linear Regression - evaluate the Model (Testing and training Accuracy, MSE for testing)

```python
# Step 4: Apply Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)

# Step 5: Evaluate the Model
Y_train_pred = linear_model.predict(X_train)
Y_test_pred = linear_model.predict(X_test)

# Calculate Mean Squared Error
mse_train = mean_squared_error(Y_train, Y_train_pred)
mse_test = mean_squared_error(Y_test, Y_test_pred)

print(f'Linear Regression - Training MSE: {mse_train}')
print(f'Linear Regression - Testing MSE: {mse_test}')
```

# Task-05: Apply KNN Regressor (Scikit-Learn) -- evaluate the Model (Testing and training Accuracy, MSE for testing)

```python
# Step 6: Apply KNN Regressor
knn_model = KNeighborsRegressor(n_neighbors=5)  # You can adjust n_neighbors as needed
knn_model.fit(X_train, Y_train)

# Step 7: Evaluate the Model
Y_train_knn_pred = knn_model.predict(X_train)
Y_test_knn_pred = knn_model.predict(X_test)

# Calculate Mean Squared Error for KNN
mse_train_knn = mean_squared_error(Y_train, Y_train_knn_pred)
mse_test_knn = mean_squared_error(Y_test, Y_test_knn_pred)

print(f'KNN Regression - Training MSE: {mse_train_knn}')
print(f'KNN Regression - Testing MSE: {mse_test_knn}')
```

# Task-06: Compare KNN & Linear Regression with the KNN Model and Linear regression as well.

```python
# Step 8: Compare KNN & Linear Regression
print("Comparison of Linear Regression and KNN Regression")
print(f'Linear Regression - Testing MSE: {mse_test}')
print(f'KNN Regression - Testing MSE: {mse_test_knn}')

# Optional: Plotting predictions
plt.scatter(Y_test, Y_test_pred, color='blue', label='Linear Regression Predictions')
plt.scatter(Y_test, Y_test_knn_pred, color='red', label='KNN Regression Predictions')
plt.xlabel('Actual Weight')
plt.ylabel('Predicted Weight')
plt.title('Actual vs Predicted Weight')
plt.legend()
plt.show()
```
